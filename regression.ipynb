```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.linear_model import SGDClassifier
import time
import pickle
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
X = 2 * np.random.rand(100, 1) # wektor pionowy
y = 4 + 3 * X + np.random.randn(100, 1)

plt.figure(figsize=(10, 6))
plt.scatter(X, y)
plt.ylim(0, 14)  # Y od 0 do 20
plt.xlim(0, 2)
plt.show()
#plt.scatter(X, y, alpha=0.7, color='blue', label='Dane')
```

### Regresja liniowa, Linear Regression, równania


$\hat{y} = \Theta_0 + \Theta_1x_1 + \Theta_2x_2 + \dots + \Theta_nx_n$        <br>
$\hat{y} = h_\Theta(x) = \Theta · x  $ <br>

#### Rozwiązanie polega na znalezieniu Θ, który minimalizuje funkcję celu (kosztu) Root Mean Square Error (RMSE) dla m instancji:



albo MSE (prościej):



```python
# dodaj stały komponent X0=1 do każdej instancji/próbki
X_b = np.c_[np.ones((100, 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)) \
.dot(X_b.T).dot(y)
print(theta_best)
```

### Regresja liniowa, predykcja: y = 4 + 3x

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new] # stały komponent d3 print(y_predict := X_new_b.dot(theta_best))

print(y_predict := X_new_b.dot(theta_best))

# Narysujmy wykres
plt.plot(X_new, y_predict, "r-", label="Predictions") # rysuje czerwoną linię predykcji
plt.plot(X, y, "b.") # rysuje niebieskie punkty danych
plt.xlabel("$x$", fontsize=18) # dodaje opis osi x
plt.ylabel("$y$", rotation=0, fontsize=18) # dodaje opis osi y
plt.legend(loc="upper left", fontsize=14) # dodaje legendę
plt.axis([0, 2, 0, 15]) # ustawia zakres osi (jak na slajdzie)
plt.show() # wyświetla wykres
```

### Linear Regression (easier)

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)

print(lin_reg.intercept_, lin_reg.coef_, "\n",
lin_reg.predict(X_new))

# to samo, co powyżej, ale prościej napisane
```
